```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#AIRCONDITIONER
audio_0 = r"C:\Users\letym\Desktop\datasets\audio\fold1\177621-0-0-120.wav"
y, sr = librosa.load(audio_0)
```

```python
fig, ax = plt.subplots()
M = librosa.feature.melspectrogram(y=y, sr=sr, n_mels = 50)
M_db = librosa.power_to_db(M, ref=np.max)
img = librosa.display.specshow(M_db, y_axis='mel', x_axis='time', ax=ax)
ax.set(title='Melspectrogram Airconditioner')
fig.colorbar(img, ax=ax, format="%+2.f dB")

plt.savefig("figure.png") # save as png
```
